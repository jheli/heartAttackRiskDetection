```python
import pandas as pd
```

Reading data

```python
df = pd.read_csv("heart_attack_prediction_dataset.csv")
df
```

Blood high preassure and blood low preassure columns as model input

```python
high_preassure = []
low_preassure = []
for bp in df['Blood Pressure'].values:
    bp_splited = bp.split('/')
    high_preassure.append(int(bp_splited[0]))
    low_preassure.append(int(bp_splited[1]))

df['Blood_Pressure_high'] = high_preassure
df['Blood_Pressure_low'] = low_preassure
```

Sex as model input

```python
sex_array = []
for sex in df['Sex']:
    if sex == 'Male':
        sex_array.append(0)
    else:
        sex_array.append(1)

df['sex_binari'] = sex_array

df['Country'].unique()

df
```

```python
diet_column = []
for d in df['Diet'].values:
    if d == 'Healthy':
        diet_column.append(1)
    elif d == 'Unhealthy':
        diet_column.append(-1)
    else:
        diet_column.append(0)

df['diet_int'] = diet_column
```
